```python
from datetime import datetime
import pandas as pd
import wget
import os
```

```python
path = f'{os.getcwd()}/PDBs'
url_download = 'https://webs.iiitd.edu.in/raghava/thpdb/thpdb.txt'
url = 'https://webs.iiitd.edu.in/raghava/thpdb/display.php?details='
```

```python
data = {
    'ID': [],
    'ThPP ID': [],
    'Therapeutic Peptide/Protein Name': [],
    'Sequence': [],
    'Functional Classification': [],
    'Molecular Weight': [],
    'Chemical Formula': [],
    'Isoelectric Point': [],
    'Hydrophobicity': [],
    'Melting Point (℃)': [],
    'Half Life': [],
    'Description': [],
    'Indication/Disease': [],
    'Pharmacodynamics': [],
    'Mechanism of Action': [],
    'Toxicity': [],
    'Metabolism': [],
    'Absorption': [],
    'Volume of Distribution': [],
    'Clearance': [],
    'Categories': [],
    'Patents Number': [],
    'Date of Issue': [],
    'Date of Expiry': [],
    'Drug Interaction': [],
    'Target': []
}
```

```python
# Download Peptide List from THPdb
filename = url_download.split('/')[len(url_download.split('/'))-1]
wget.download(url_download)
os.system(f'mv {filename} PDBs/THPdb_releases/{filename}')
```

```python
with open(f'{path}/THPdb_releases/{filename}', encoding='utf8') as file:
    lines = file.readlines()
    for line in lines:
        for pep_info in line.split('\n'):
            #ignore the blank lines
            if not pep_info.strip():
                continue
            index = 0
            for key in data.keys():
                data[key].append(pep_info.split('\t')[index])
                index += 1
```

```python
df = pd.DataFrame(data)
```

```python
#Add links to peptides in THPdb
links = []
for pep_id in df['ID']:
    links.append(f'{url}{pep_id}')
df['Link'] = links
```

```python
dt = datetime.now()
df.to_csv(f'{path}/THPdb_releases/THPdb_{dt.strftime("%Y_%m_%d")}.csv', index=False)
df.to_csv(f'{path}/THPdb.csv', index=False)
```
