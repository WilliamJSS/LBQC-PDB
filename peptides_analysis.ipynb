```python
# Import R packages
rpep = importr('Peptides')
rtdv = importr('tidyverse')
rggp = importr('ggpubr')
```

```python
# Read LBQCPDB.csv
df = pl.read_csv(f'{path}/LBQCPDB.csv')
sequences = df.select(pl.col('Sequence')).limit(10).rows()
```

```python
# Molecular Weight and Charge correlation
list_of_bytes = [bytes(element, 'utf-8') for element in sequences[0]]
b = b''.join(list_of_bytes)
mz = rpep.mz(b)
```

```python
# Charge
charge = rpep.charge(sequences[0], pH = 7, pKscale = "Lehninger")
```

```python
# Hidrophobicity
hp = rpep.hydrophobicity(sequences, scale = "KyteDoolittle")
```

```python
# Molecular Weight
mw = rpep.mw(
  sequences,
  monoisotopic = False,
  avgScale = "expasy",
  label = "none",
  aaShift = null
)
```
