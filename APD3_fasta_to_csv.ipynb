```python
from datetime import datetime
import pandas as pd
```

```python
path = 'D:\WilliamJSS\Projects\LBQC\LBQC-PDB\PDBs\APD3'
```

```python
data = {
    'APD_ID': [],
    'NAME': [],
    'SEQUENCE': [],
    'CLASS': []
}
```

```python
with open(f'{path}/APD_sequence_release_09142020.fasta') as file:
    lines = file.readlines()
    for i in range(0, len(lines), 2):
        line_info = lines[i].replace(' (','|').replace(')','|').split('|')
        line_seq = lines[i+1].split('\n')
        data['APD_ID'].append(int(line_info[0][1:]))
        data['NAME'].append(line_info[1])
        data['SEQUENCE'].append(line_seq[0])
        data['CLASS'].append(line_info[2])
```

```python
df = pd.DataFrame(data)
```

```python
dt = datetime.now()
df.to_csv(f'{path}/APD3_{dt.strftime("%Y_%m_%d")}.csv', index=False)
```
