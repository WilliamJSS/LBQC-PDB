```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from datetime import datetime
import pubmed as pm
import pandas as pd
import time
import os
```

```python
path = f'{os.getcwd()}/PDBs'
url = 'https://webs.iiitd.edu.in/raghava/fermfoodb/display_sub.php?details=FMDB'
```

```python
data = {
    'Title': [],
    'Peptide Sequence': [],
    'Length': [],
    'Food_Matrix': [],
    'Protein Name': [],
    'pH': [],
    'Temperature': [],
    'Incubation Time': [],
    'Activity': [],
    'Experiment': [],
    'Model': [],
    'Assay for Activity Measurement': [],
    'Starter Culture': [],
    'Hydrolysis': [],
    'Method of Analysis': [],
    'M/Z ratio': [],
    'Mass': [],
    'IC50': []
}
```

```python
driver = webdriver.Chrome()
```

```python
pep_id = 1
while pep_id > 0:
    driver.get(f'{url}{pep_id}')
    time.sleep(2)
    try:
        for i in range(5, 23):
            key = driver.find_element(By.XPATH, f'/html/body/div[3]/main/div/div/table[2]/tbody/tr[{i}]/td[1]/b/font')
            value = driver.find_element(By.XPATH, f'/html/body/div[3]/main/div/div/table[2]/tbody/tr[{i}]/td[2]/font')
            data[key.text].append(value.text) 
        pep_id = pep_id + 1
    except:
        print(f'FermFooDB_ID FMDB{pep_id} not found')
        pep_id = 0
```

```python
df = pd.DataFrame(data)
```

```python
#Add links to peptides in FermFooDb
links = []
for i in range(1, len(df)+1):
    links.append(f'{url}{i}')
df['Link'] = links
```

```python
# Get DOI from PubMed
doi_list = []

for link in df['Link']:
    driver.get(link)
    pmid = driver.find_element(By.XPATH, '/html/body/div[3]/main/div/div/table[2]/tbody/tr[3]/td[2]/font')
    doi_list.append(pm.get_doi_from_pmid(driver, pmid.text))
    
df['DOI'] = doi_list
```

```python
driver.close()
```

```python
dt = datetime.now()
df.to_csv(f'{path}/FermFooDb_releases/FermFooDb_{dt.strftime("%Y_%m_%d")}.csv', index=False)
df.to_csv(f'{path}/FermFooDb.csv', index=False)
```
